```python
#select from database mongolian text
mycursor = mydb.cursor()

mycursor.execute("SELECT text FROM mongolian_text_sentiment_analysis.mongolian_text_with_emoji")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
#concatanate all mongolian text
mylist = []
for x in myresult:
    mylist = mylist + list(x)
```

```python
import googletrans
from googletrans import Translator
```

```python
#function to translate mongolian text into english
def translate_all() : 
    final_translated_list = []
    #send 100 requests to translate
    for i in range(100):
        #join text by 10 using a delimiter ==> to save number of request to Google Translate
        text = mylist[i*10:(i+1)*10]
        joined_text = '==='.join(text)
        translator = Translator()
        translated_text = translator.translate(joined_text, src='mn', dest='en')
        #split text using delimiter to find out the original version of the text
        if __name__ == '__main__':
            translated_list = translated_text.text.split('===')
            final_translated_list = final_translated_list + translated_list
    return final_translated_list
```

```python
#create new table to save translated text
mycursor.execute("DROP TABLE IF EXISTS mongolian_text_sentiment_analysis.translated_text")
mycursor.execute("CREATE TABLE mongolian_text_sentiment_analysis.translated_text (id INT AUTO_INCREMENT PRIMARY KEY, text LONGTEXT)")
```

```python
#execute the function to translate and insert the result into the newly created table
final_translated_list = translate_all()
for translation in final_translated_list : 
    sql = "INSERT INTO mongolian_text_sentiment_analysis.translated_text (text) VALUES (%s)"
    val = translation
    mycursor.execute(sql, (val,))
    mydb.commit()
```

```python
#join the table with mongolian text with the one with translated text to create a new table named Result
mycursor.execute("CREATE TABLE mongolian_text_sentiment_analysis.result AS (SELECT mongolian_text_sentiment_analysis.mongolian_text_with_emoji.index as id, mongolian_text_sentiment_analysis.mongolian_text_with_emoji.text as mongolian_text, mongolian_text_sentiment_analysis.translated_text.text as english_text, mongolian_text_sentiment_analysis.mongolian_text_with_emoji.emoji as emoji FROM   mongolian_text_sentiment_analysis.mongolian_text_with_emoji JOIN mongolian_text_sentiment_analysis.translated_text ON mongolian_text_sentiment_analysis.mongolian_text_with_emoji.index = mongolian_text_sentiment_analysis.translated_text.id )")
```
