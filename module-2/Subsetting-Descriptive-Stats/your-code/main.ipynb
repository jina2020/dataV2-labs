#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here orders[orders.amount_spent > 50]['CustomerID']
orders_bis = orders[['CustomerID', 'Quantity']]
orders_bis = orders_bis.groupby(['CustomerID']).sum()
orders_bis[orders_bis.Quantity > 50]
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
orders_bis = orders[['CustomerID', 'Quantity', 'Country']]
orders_bis = orders_bis.groupby(['CustomerID', 'Country']).sum().reset_index()
orders_bis[(orders_bis.Quantity > 50) & (orders_bis.Country == 'Spain')]
```

#### What's the  average amount of money spent by country?

```python
# your code here
orders_average_amount_by_country = orders[['amount_spent', 'Country']]
orders_average_amount_by_country = orders_average_amount_by_country.groupby(['Country']).sum() / orders_average_amount_by_country.groupby(['Country']).count()
orders_average_amount_by_country
```

#### What is the average amount spent per year?

```python
# your code here
orders_average_amount_by_year = orders[['amount_spent', 'year']]
orders_average_amount_by_year = orders_average_amount_by_year.groupby(['year']).sum() / orders_average_amount_by_year.groupby(['year']).count()
orders_average_amount_by_year
```
