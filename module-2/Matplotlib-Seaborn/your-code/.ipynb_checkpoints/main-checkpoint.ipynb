#### Plot (x, y) and (x, z).

```python
# Method 1
plt.plot(x, y, label='linear')
plt.plot(x, z, label='quadratic')
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT X, Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,z)
ax2.set_title('QUADRATIC PLOT X, Z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize = (15, 5))
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT X, Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,z)
ax2.set_title('QUADRATIC PLOT X, Z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = [math.exp(i) for i in x]

fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize = (15, 5))
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT X, Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,z)
ax2.set_title('EXPONENTIAL PLOT X, Z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_yscale('log')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_step_month = pd.DataFrame()
median_step_month["Steps"] = fitbit.groupby(["Months_encoded"])["Steps"].median()
median_step_month.reset_index(inplace = True)
plt.plot(median_step_month["Months_encoded"], median_step_month["Steps"], label = "Monthly median step")
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_step_work_weekend = pd.DataFrame()
median_step_work_weekend["Steps"] = fitbit.groupby(["Work_or_Weekend"])["Steps"].median()
median_step_work_weekend.reset_index(inplace = True)
plt.plot(median_step_work_weekend["Work_or_Weekend"], median_step_work_weekend["Steps"], label = "Weekly median step")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fitbit_list = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
for x in fitbit_list: 
    plt.scatter(fitbit[x], fitbit["Steps"], label = x + " VS Steps")
```
